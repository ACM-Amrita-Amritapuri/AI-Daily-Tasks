```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dt=pd.read_csv("https://raw.githubusercontent.com/ACM-Amrita-Amritapuri/AI-Daily-Tasks/main/Data%20Pre-Processing%20(8-11-22)/Data_Set.csv")
```

```python
print("4 features are available")
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
dt['ObservationDate']= label_encoder.fit_transform(dt['ObservationDate'])
dt['Country/Region']= label_encoder.fit_transform(dt['Country/Region'])
dt['Last Update']= label_encoder.fit_transform(dt['Last Update'])
```

```python
plt.figure(figsize=(10,10))
cor = dt.corr()
sns.heatmap(data = cor, annot =True)
```
