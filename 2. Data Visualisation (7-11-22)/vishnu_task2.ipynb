```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
house_filepath ='/content/California_Houses.csv'
house=pd.read_csv(house_filepath, parse_dates=True)
```

```python
house
```

```python
#qs1
sns.lineplot(data=house['Tot_Rooms'],label="total rooms")
sns.lineplot(data=house['Tot_Bedrooms'],label="total bedrooms")
```

```python
#qs2
sns.barplot(x=house.Median_Age, y=house.query('Median_Income<6').Median_Income)
```

```python
#qs3
corr=house.corr()
sns.heatmap(corr)
```

```python
#qs4
sns.scatterplot(x=house['Median_Income'],y=house['Median_House_Value'],hue=house['Median_Age'])
```

```python
#qs5
sns.kdeplot(data=house, x=house.Distance_to_LA, hue=house.query('30<=Median_Age<=40').Median_Age )
```
